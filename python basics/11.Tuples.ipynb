#### Packing and Unpacking tuple
Packing is the process of grouping multiple values into a single tuple. This happens automatically when you assign multiple comma-separated values to a single variable. Parentheses are optional but often used for clarity.

Unpacking is the process of extracting individual values from a tuple and assigning them to separate variables. This requires the number of variables on the left-hand side of the assignment to match the number of elements in the tuple on the right-hand side. 

```python
## packing

packed_tuple=1,"Hello",3.14
print(packed_tuple)
```

```python
##unpacking a tuple

a,b,c=packed_tuple

print(a)
print(b)
print(c)
```

```python
## Unpacking with *
numbers=(1,2,3,4,5,6)
first,*middle,last=numbers
print(first)
print(middle)
print(last)
```

```python
## Nested Tuple
## Nested List
lst=[[1,2,3,4],[6,7,8,9],[1,"Hello",3.14,"c"]]
print(type(lst))
lst[0][0:3]
```

```python
lst=[[1,2,3,4],[6,7,8,9],(1,"Hello",3.14,"c")]
print(type(lst[2]))
lst[2][0:3]
```

```python
nested_tuple = ((1, 2, 3), ("a", "b", "c"), (True, False))

## access the elements inside a tuple
print(nested_tuple[0])
print(nested_tuple[1][2])
```

```python
## iterating over nested tuples
for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item,end=" ")
    print()
```
